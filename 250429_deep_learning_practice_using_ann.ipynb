## 모델링

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = df.iloc[:,0:8]
y = df.iloc[:,8]
```

```python

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu', name='Dense_1'))
model.add(Dense(8, activation='relu', name='Dense_2'))
model.add(Dense(1, activation='sigmoid',name='Dense_3'))
model.summary() # ppt14 딥러닝과 다중분류문제 해결 18p 참고하기
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history=model.fit(X, y, epochs=100, batch_size=5)
```

# 붓꽃 다중 분류

## 데이터 불러오기

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
y
```

```python
len(X)
```

```python
df = pd.DataFrame(data= iris.data, columns=iris.feature_names)
df["species"]=iris.target
df.rename(columns={'sepal length (cm)' : 'sepal_length', 'sepal width (cm)' : 'sepal_width','petal length (cm)' : 'petal_length', 'petal width (cm)' : 'petal_width' }, inplace=True)
df
```

## 상관관계 분석

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df, hue='species'); # 전체 상관도 확인
plt.show()
```

## 원핫인코딩

```python
# 속성을 X, 클래스를 y로 저장
X = df.iloc[:,0:4]
y = df.iloc[:,4]

print(X[0:5])
print(y[0:5])
```

```python
# 원핫 인코딩 처리
y = pd.get_dummies(y) # 사실 할 필요 없음
print(y[0:5])
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# 모델 설정
model = Sequential()
model.add(Dense(12,  input_dim=4, activation='relu'))
model.add(Dense(8,  activation='relu'))
model.add(Dense(3, activation='softmax'))
model.summary()

# 모델 컴파일
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 모델 실행
history=model.fit(X, y, epochs=30, batch_size=5)
```
